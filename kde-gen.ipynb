```python
# splitvars:
# Age
# Sex
# Risk of Drug Abuse
# Manual QC
# Task
# Run - 'bids_meta.session_id'?

splitvars = [ 'bids_meta__Manufacturer', 'bids_meta__ManufacturersModelName', 'bids_meta__modality', 'bids_meta__TaskName' ]

df[splitvars].describe()
```

```python
def kdegen(dataframe, x_list):
    # returns x and y as dictionary
    
    # remove NaNs
    x_list = x_list[~np.isnan(x_list)]
    kernel = stats.gaussian_kde(x_list)
    x = np.linspace(x_list.min(), x_list.max(), num=1000)
    y = kernel.evaluate(dataframe[x_list])
    return(list(zip(x, y)))

# FIXME: add mean, median etc to this (make it a dict with mean, median, ..., kde)
def kdetuples(dataframe):
    kdedict = {}
    for iqm in iqms:
        kdedict[iqm] = kdegen(dataframe, iqm)
    return(kdedict)
```

```python
def subsetdf(dataframe, varname, varval):
    return dataframe.loc[dataframe[varname] == varval]

# Logic to identify the n of the subgroup
def subgroupsize(subdf):
    return(len(subdf.iloc[:,1]))

def writejson(data, filename):
    with open('/abcdqc_data/batchserver/output/' + filename, 'w') as outfile:
        json.dump(data, outfile)

def sv_iter(dataframe, setname):
    # iterate over splitvars
    for sv in splitvars:
        # loop over all possible values of the splitvar sv
        for varvalue in dataframe[sv].unique():
            # get subset of df
            subdf = subsetdf(dataframe, sv, varvalue)
            # skip groups with size < 100 for confidentiality reasons
            if subgroupsize(subdf) < 100:
                continue
            # write KDE tuples (kdedict) for subdf to json
            writejson(kdetuples(subdf), setname + '-' + sv + '-' + varvalue + '.json')

# lists of modality-based dataframes and IQM lists
df_mods = [ 'T1w', 'T2w', 'bold']
iqmlist = [ 't1wiqms', 't2wiqms', 'boldiqms']

# divide df into 3 sets based on modality because they contain different sets of IQMs
start_time = time.time()
for mrimode in df_mods:
    # use a dictionary to allow different variable names as mentioned here:
    # https://stackoverflow.com/questions/6181935/how-do-you-create-different-variable-names-while-in-a-loop
    d = dict() # reset dictionary to only have one entry
    d[mrimode] = subsetdf(df, 'bids_meta__modality', mrimode)
    # d.values()[0] returns the first entry in the dictionary
    mode_df = list(d.values())[0]
    # iterate over all 3 subsets
    sv_iter(mode_df, mrimode)
print(time.time() - start_time, " seconds")
```
